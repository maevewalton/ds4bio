hypothesis.

The reason the research was conducted is because nation-wide, bumble bees are experiencing die-offs, a plumet in the population, with no underlying cause. This is known as Colony Collaspe Disorder (CDC). By investigating the different bumble bee populations in Alaska, scientist are hoping to reduce the genetic diversity of all North American bumble bees.

It was hypothesized that Delta Junction would have the most diverse and abundant bumble bee population as it is the most rural agricultural location.

By also investingating the nemotode parasitic impact on bumble bees, the researchers hope to decrease the loss of colonies that have a presence of nemotodes.

Do certain species of bumble bee fair better against parasites?
